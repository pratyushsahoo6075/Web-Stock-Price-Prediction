```python
import matplotlib.pyplot as plt

# Define the plot_graph function to accept figure size, data, and column name
def plot_graph(figsize, data, column_name):
    plt.figure(figsize=figsize)  # Set the figure size
    plt.plot(data)  # Plot the data
    plt.xlabel("Index")  # Label for x-axis
    plt.ylabel(column_name)  # Label for y-axis
    plt.title(f"{column_name} Over Time")  # Title of the plot
    plt.show()  # Display the plot

# Example usage with google_data DataFrame
for column in google_data.columns:
    plot_graph((15, 5), google_data[column], column)
```

```python
Adj_Close_price=google_data[['Adj Close']]
```

```python
max(Adj_Close_price.values),min(Adj_Close_price.values)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(Adj_Close_price)
scaled_data
```

```python
len(scaled_data)
```

```python
x_data=[]
y_data=[]

for i in range(100,len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])
    
import numpy as np
x_data[0],y_data[0]
```

```python
int(len(x_data)*0.7)
```

```python
4980-100-int(len(x_data)*0.7)
```

```python
splitting_len = int(len(x_data)*0.7)
x_train=x_data[:splitting_len]
y_train=y_data[:splitting_len]

x_test=x_data[splitting_len:]
y_test=y_data[splitting_len:]
```

```python
# Assuming x_train, y_train, x_test are lists
x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)

# Now you can use the .shape attribute
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(x_train,y_train,batch_size=1,epochs= 2)
```

```python
 model.summary()
```

```python
prediction = model.predict(x_test)
```

```python
inv_predictions=scaler.inverse_transform(prediction)
inv_predictions
```

```python
inv_y_test=scaler.inverse_transform(y_test)
inv_y_test
```

```python
import numpy as np
# Assuming inv_predictions and inv_y_test are numpy arrays
rmse = np.sqrt(np.mean((inv_predictions - inv_y_test) ** 2))
print(rmse)
```

```python
import pandas as pd

# Ensure inv_y_test and inv_predictions are numpy arrays and reshaped correctly
inv_y_test = inv_y_test.reshape(-1)
inv_predictions = inv_predictions.reshape(-1)

# Calculate the index range to match the length of the data
index_range = google_data.index[splitting_len + 100: splitting_len + 100 + len(inv_y_test)]

# Create the DataFrame with the correct index
plotting_data = pd.DataFrame(
    {
        'original_test_data': inv_y_test,
        'predictions': inv_predictions
    },
    index=index_range
)

# Display the first few rows of the DataFrame
print(plotting_data.head())
```

```python
import matplotlib.pyplot as plt

# Create the plot
plt.figure(figsize=(15, 6))
plt.plot(plotting_data.index, plotting_data['original_test_data'], label='Original Test Data')
plt.plot(plotting_data.index, plotting_data['predictions'], label='Predictions', linestyle='--')
plt.xlabel('Years')
plt.ylabel('test data')
plt.title(' test data of Google Data')  # Set the title here
plt.legend()
plt.show()
```
